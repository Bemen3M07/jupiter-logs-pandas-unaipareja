```python
import logging

# Configuració del logger principal
logger = logging.getLogger("MainLogger")
logger.setLevel(logging.DEBUG)

# Handler per guardar missatges d'info i superiors
info_handler = logging.FileHandler("info.log")
info_handler.setLevel(logging.INFO)

# Handler per guardar missatges d'error i superiors
error_handler = logging.FileHandler("error.log")
error_handler.setLevel(logging.ERROR)

# Formatador comú
formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")
info_handler.setFormatter(formatter)
error_handler.setFormatter(formatter)

# Afegim els handlers al logger principal
logger.addHandler(info_handler)
logger.addHandler(error_handler)

# Exemple de log per comprovar que funciona
logger.info("Aquest és un missatge d'info.")
logger.error("Aquest és un missatge d'error.")

# Handler per mostrar missatges per pantalla
console_handler = logging.StreamHandler()
console_handler.setLevel(logging.INFO)
console_handler.setFormatter(formatter)

# Afegim el handler de pantalla al logger principal
logger.addHandler(console_handler)

# Exemple de log per comprovar que funciona
logger.info("Missatge d'info que també es mostra per pantalla.")
logger.error("Missatge d'error que també es mostra per pantalla.")

# Logger especial per a CSV
csv_logger = logging.getLogger("CSVLogger")
csv_logger.setLevel(logging.INFO)

# Handler per generar logs en format CSV
csv_handler = logging.FileHandler("logs.csv", mode="w", encoding="utf-8")
csv_formatter = logging.Formatter("%(asctime)s,%(levelname)s,%(message)s")
csv_handler.setFormatter(csv_formatter)

# Afegim el handler al logger especial
csv_logger.addHandler(csv_handler)

# Exemple de log per comprovar que funciona
csv_logger.info("Missatge d'info en format CSV.")
csv_logger.error("Missatge d'error en format CSV.")

def generate_logs():
    for i in range(1, 21):
        if i % 2 == 0:
            logger.info(f"El número {i} és parell.")
            csv_logger.info(f"CSV Log: El número {i} és parell.")
        else:
            logger.error(f"El número {i} és senar!")
            csv_logger.error(f"CSV Log: El número {i} és senar!")

generate_logs()
```
