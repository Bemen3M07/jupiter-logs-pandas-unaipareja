# Exercici 6
De la mateixa manera que a l'exercici 5 caldrà investigar una llibreria per a realitzar gràfics de dades visuals i elaborar com a mínim el codi necessari per presentar de forma que tinguin sentit amb l'àmbit del problema.
Com a restricció caldrà com a mínim generar 4 gràfiques coherents.
Es pot utilitzar la llibreria python matploit

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar el fitxer CSV
df = pd.read_csv('notes.csv')

# Crear la columna "Mitjana" per cada alumne (mitjana de les notes M01, M04, M03, M05)
df['Mitjana'] = df[['M01', 'M04', 'M03', 'M05']].mean(axis=1)

# 1. Gràfic de barres de la mitjana de nota per alumne
plt.figure(figsize=(10, 6))
plt.bar(df['Alumne'], df['Mitjana'], color='skyblue')
plt.title('Mitjana de Nota per Alumne')
plt.xlabel('Alumne')
plt.ylabel('Nota Mitjana')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Gràfic de sectors per percentatge d'aprovats i suspesos
aprovats = (df['Mitjana'] >= 5).sum()
suspesos = (df['Mitjana'] < 5).sum()
labels = ['Aprovats', 'Suspesos']
sizes = [aprovats, suspesos]
colors = ['lightgreen', 'salmon']

plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, explode=(0.1, 0))
plt.title('Percentatge d\'Aprovats i Suspensió')
plt.axis('equal')  # Perquè el gràfic sigui un cercle perfecte
plt.show()

# 3. Gràfic de línia per mostrar l'evolució de les mitjanes en funció de l'alumne
plt.figure(figsize=(10, 6))
plt.plot(df['Alumne'], df['Mitjana'], marker='o', color='orange', linestyle='-', linewidth=2, markersize=8)
plt.title('Evolució de les Mitjanes de Nota per Alumne')
plt.xlabel('Alumne')
plt.ylabel('Nota Mitjana')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 4. Gràfic de barres per mostrar la mitjana per assignatura (M01, M04, M03, M05)
# Es calcularà la mitjana per cada assignatura
mitjanes_assignatures = df[['M01', 'M04', 'M03', 'M05']].mean()

plt.figure(figsize=(10, 6))
mitjanes_assignatures.plot(kind='bar', color='lightblue')
plt.title('Mitjana de Nota per Assignatura')
plt.xlabel('Assignatures')
plt.ylabel('Mitjana de Nota')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
